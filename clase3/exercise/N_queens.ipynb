## RepresentaciГіn del problema en Python

Para trabajar con el problema de las N reinas, necesitamos una forma sencilla y eficiente de representar el tablero y la ubicaciГіn de las reinas.

La forma mГЎs comГәn y prГЎctica es usar una lista de tamaГұo N, donde:
- El Гӯndice de la lista representa la columna del tablero.
- El valor almacenado en ese Гӯndice representa la fila donde estГЎ ubicada la reina en esa columna.

Por ejemplo, para *N=8*, un arreglo como este:

```python
board = [0, 4, 7, 5, 2, 6, 1, 3]
```

Significa:
- La reina en la columna 0 estГЎ en la fila 0.
- La reina en la columna 1 estГЎ en la fila 4.
- La reina en la columna 2 estГЎ en la fila 7.
- ... y asГӯ sucesivamente.

De esta forma podemos representar diferentes estados y empezar a "mover" a las reinas. Pero antes, necesitamos una funciГіn que inicialice un estado al azar:

```python
def initialize_random_state(n: int = 8):
    """
    Initializes a random state for the N-queens problem.
    Returns a list where each index represents a column
    and the value is the row where the queen is placed.
    """
    return [random.randint(0, n - 1) for _ in range(n)]

# Ejemplo de uso:
estado_inicial = initialize_random_state()
print("Estado inicial aleatorio:", estado_inicial)
```

## GeneraciГіn de vecinos

En los algoritmos de bГәsqueda local, la idea es moverse en el espacio de soluciones explorando vecinos del estado actual.

**ВҝQuГ© es un vecino en el problema de las N reinas?**

Dado un estado (una configuraciГіn particular de reinas en el tablero), un vecino es otro estado que resulta de hacer una pequeГұa modificaciГіn en el estado actual.

En nuestro caso, donde cada reina estГЎ en una columna, una forma natural de definir vecinos es:

1. Elegir una columna.
2. Mover la reina de esa columna a una fila diferente dentro de la misma columna.

**Ejemplo**

Si el estado actual es:

```python
board = [0, 4, 7, 5, 2, 6, 1, 3]
```

Un vecino puede ser:

`Mover la reina en la columna 2 (que estГЎ en fila 7) a la fila 0`, dando:

```python
neig_board = [0, 4, 0, 5, 2, 6, 1, 3]
```

A continuaciГіn, definamos una funciГіn que genera todos los vecinos posibles a partir de un estado dado:

```python
def generate_neighbors(state):
    """
    Generates all neighboring states from the current state.
    A neighbor is a state where one queen is moved to a different
    row within its column.
    
    Args:
        state (list): List representing the queen positions.
        
    Returns:
        list: List of neighboring states.
    """
    n = len(state)
    neighbors = []
    
    for col in range(n):
        current_row = state[col]
        for new_row in range(n):
            if new_row != current_row:
                neighbor = state.copy()
                neighbor[col] = new_row
                neighbors.append(neighbor)
                
    return neighbors

# Ejemplo de uso:
estado_actual = [0, 4, 7, 5, 2, 6, 1, 3]
vecinos = generate_neighbors(estado_actual)
print(f"Cantidad de vecinos generados: {len(vecinos)}")
print("Algunos vecinos:")
for vecino in vecinos[:5]:
    print(vecino)
```

**ВҝQuГ© hace esta funciГіn?**

1. Para cada columna, prueba mover la reina a todas las filas distintas de la actual.
2. Copia el estado actual y modifica la posiciГіn para crear un vecino nuevo.
3. Devuelve una lista con todos los vecinos posibles.

## FunciГіn de costo

En los algoritmos de bГәsqueda local, cada estado (o soluciГіn parcial) tiene un *costo* que indica quГ© tan lejos estГЎ de ser una soluciГіn vГЎlida.

En el caso del problema de las N reinas, una soluciГіn vГЎlida es aquella donde ninguna reina puede atacar a otra. Por lo tanto, una forma natural de definir la funciГіn de costo es:

> Sea:
> - N el tamaГұo del tablero (y cantidad de reinas)
> - $s = [s_0, s_1, ..., s_{N-1}]$ un estado, donde $s_i \in \{ 0, 1, ..., N-1 \}$ representa la fila en la quie estГЎ la reina de la columna *i*.
>
> Entonces, la funciГіn de costo $C(s)$ mide el nГәmero total de pares de reinas que se atacan entre sГӯ:
>
> $$C(s) = \sum_{i=0}^{N-1} \sum_{j=i+1}^{N-1} \delta(s_i, s_j)$$
> donde la funciГіn $\delta(s_i, s_j)$ se define como:
> $$ \delta(s_i, s_j) = \begin{cases} 1 & \text{si } s_i = s_j \quad \text{(misma fila)} \\ 1 & \text{si } |s_i - s_j| = |i - j| \quad \text{(misma diagonal)} \\ 0 & \text{en otro caso}\end{cases}  $$

Queremos minimizar esta funciГіn:
- Si el costo es `0`, ВЎtenemos una soluciГіn vГЎlida!
- Si el costo es mayor que `0`, hay conflictos que resolver.

ВҝCuГЎndo dos reinas se atacan? Dos reinas se atacan si estГЎn:

- En la misma fila
- En la misma diagonal (ascendente o descendente)
- вҡ пёҸ No necesitamos preocuparnos por columnas, porque la representaciГіn del problema (una reina por columna) ya garantiza que no haya mГЎs de una reina por columna.

La implementaciГіn en Python es:

```python
def cost(state):
    """
    Calculates the number of pairs of queens that are attacking each other.
    
    Args:
        state (list): A list where state[i] is the row of the queen in column i.
        
    Returns:
        int: Total number of conflicting pairs.
    """
    n = len(state)
    conflicts = 0
    for i in range(n):
        for j in range(i + 1, n):
            same_row = state[i] == state[j]
            same_diag = abs(state[i] - state[j]) == abs(i - j)
            if same_row or same_diag:
                conflicts += 1
    return conflicts

# Ejemplo de uso
state = [0, 4, 7, 5, 2, 6, 1, 3]  # SoluciГіn valida para N=8
print("Estado:", state) 
print("Costo:", cost(state))     

state = [0, 4, 7, 5, 2, 6, 1, 0]
print("Estado:", state)
print("Costo:", cost(state))
```

Esta funciГіn se usa para:
- Evaluar quГ© tan buena es una soluciГіn actual.
- Comparar varios vecinos y elegir el de menor costo.
- Saber si ya llegamos a una soluciГіn (cuando el costo es 0).

## VisualizaciГіn del tablero

Una forma muy Гәtil de comprender y depurar soluciones al problema de las N reinas es poder visualizar el estado del tablero. Para esto, podemos imprimirlo en consola usando caracteres ASCII, representando las casillas vacГӯas y las reinas.

ВҝCГіmo lo representamos? Usamos los siguientes sГӯmbolos:

- `вҷӣ` para indicar una reina
- `|` y `---` para simular los bordes del tablero

Definamos la funciГіn que nos permita armar la representaciГіn:

```python
def print_board_ascii(state: list):
    """
    Prints an ASCII representation of the N-Queens board.
    
    Each index in the input list represents the column,
    and the value at that index represents the row where the queen is placed.

    Args:
        state (list): A list of integers where state[i] is the row of the queen in column i.

    Example:
        For board = [0, 4, 7, 5, 2, 6, 1, 3], prints a visual 8x8 board with queens.
    """
    n = len(state)
    horizontal_border = "+" + "---+" * n
    for row in range(n):
        print(horizontal_border)
        row_str = ""
        for col in range(n):
            if state[col] == row:
                row_str += "| вҷӣ "
            else:
                row_str += "|   "
        row_str += "|"
        print(row_str)
    print(horizontal_border)

# Ejemplo de uso
state = [0, 4, 7, 5, 2, 6, 1, 3]  # SoluciГіn vГЎlida para N=8
print_board_ascii(state)
```

## Algoritmo de gradiente descendente (hill climbing)

Tal como vimos en los videos, este algoritmo parte de un estado inicial y se mueve siempre al mejor vecino, es decir, al estado vecino con menor cantidad de conflictos (menor costo), esperando llegar a una soluciГіn sin conflictos.

**Estructura general del algoritmo**
1. Generar un estado inicial aleatorio.
2. Evaluar su costo.
3. Generar todos sus vecinos.
4. Elegir el vecino con menor costo.
5. Si es mejor que el actual, avanzar a ese estado.
6. Si no hay mejora posible вҶ’ nos quedamos en un Гіptimo local.

Veamos la implementaciГіn:

```python
N = 8 # NГәmero de reinas
MAX_STEPS = 1000 # NГәmero maximo de iteraciones

def grad_desc(n, max_steps=1000) -> tuple:
    """
    Hill Climbing algorithm for solving the N-Queens problem.
    
    Args:
        n (int): Size of the board (number of queens).
        max_steps (int): Maximum number of steps before giving up.
        
    Returns:
        tuple: (final_state, final_cost)
    """

    # 1. Inicializamos un estado al azar
    current = initialize_random_state(n)
    # Calculamos el costo
    current_cost = cost(current)

    print("Estado inicial:", current)
    print("Costo:", current_cost)
    print_board_ascii(current)
    print("------------")
    
    last_step = max_steps
    for step in range(max_steps):
        # 3. Generamos todos sus vecinos.
        neighbors = generate_neighbors(current)
        # Calculamos todos los costos de los vecinos
        neighbor_costs = [cost(neighbor) for neighbor in neighbors]
        # Obtenemos el minimo costo
        min_cost = min(neighbor_costs)

        # Si todos los vecinos tienen el mismo valor o mayor, terminamos.
        if min_cost >= current_cost:
            last_step = step
            break

        # Elegimos el vecino con menor costo.
        best_index = neighbor_costs.index(min_cost)
        current = neighbors[best_index]
        current_cost = min_cost
        

    if current_cost == 0:
        print("вң… ВЎSe encontrГі una soluciГіn!")
    else:
        print("вқҢ No se encontrГі una soluciГіn :(")

    print("Iteraciones ejecutadas:", last_step)
    print("Estado final:", current)
    print("Costo final:", current_cost)
    print_board_ascii(current)

    return current, current_cost
```

### Simmulated Annealing

Llegamos a la parte que **vos debГ©s desarrollar**. Ahora te toca implementar el algoritmo de bГәsqueda Simulated Annealing.

Para eso, debГ©s completar la funciГіn que aparece mГЎs abajo.

> вҡ пёҸ No modifiques los argumentos ni el tipo de salida: la funciГіn debe recibir exactamente los parГЎmetros indicados y devolver el resultado especificado.

```python
import math

def simulated_annealing(n: int, initial_temp: float, cooling_rate: float, max_steps: int = 1000) -> tuple:
    """
    Simulated Annealing algorithm for solving the N-Queens problem.
    
    Args:
        n (int): Size of the board (number of queens).
        initial_temp (float): Starting temperature.
        cooling_rate (float): How much the temperature decreases per step.
        max_steps (int): Maximum number of iterations.
    
    Returns:
        tuple: (final_state, final_cost)
    """
    current = initialize_random_state(n)
    current_cost = cost(current)
    temperature = initial_temp

    print("Estado inicial:", current)
    print("Costo:", current_cost)
    print_board_ascii(current)
    print("------------")

    last_step = 0

    for step in range(max_steps):
        
        # Si encontramos una soluciГіn, salimos
        if current_cost == 0:                               
            break

        # 1. Generar un vecino RANDOM (no todos)
        neighbors = generate_neighbors(current)
        next_state = random.choice(neighbors)
        next_cost = cost(next_state)

        delta = next_cost - current_cost                                # Calculamos la diferencia de costo entre el vecino y el estado actual

        # 2. Decidir si aceptar
        if delta < 0:                                                   # Si el vecino es mejor, lo aceptamos siempre
            current = next_state
            current_cost = next_cost                                    
        else:                                                           # Si es peor, lo aceptamos con una probabilidad que depende de la temperatura        
            prob = math.exp(-delta / temperature)
            if random.random() < prob:
                current = next_state
                current_cost = next_cost

        # 3. Enfriar temperatura
        temperature *= cooling_rate

        last_step = step + 1                                            # Si no se encontrГі una soluciГіn, sumamos 1 a la cantidad de iteraciones ejecutada

        if temperature < 1e-8:                                          # Si la temperatura es muy baja, salimos
            break
        
    if current_cost == 0:
        print("вң… ВЎSe encontrГі una soluciГіn!")
    else:
        print("вқҢ No se encontrГі una soluciГіn :(")

    print("Iteraciones ejecutadas:", last_step)
    print("Estado final:", current)
    print("Costo final:", current_cost)
    print_board_ascii(current)
    print("------------")

    return current, current_cost 
```
